```julia
# order of the polynomial to use
k = 3
```

```julia
# fit using a function of the form f(x) = u1 x^k + u2 x^(k-1) + ... + uk x + u{k+1}
n = length(raw[:,1])

# create the A matrix with coefficients
# the main objective here is to find the coefficients
# doing raw[i,1]^(k+1-j) is simply like doing x^i.
A = zeros(n,k+1)
for i = 1:n
    for j = 1:k+1
        A[i,j] = raw[i,1]^(k+1-j)
    end
end
```

```julia
# use Gurobi and Julia
using JuMP, Gurobi
```

```julia
m = Model(solver=GurobiSolver(OutputFlag=0))

# declare variable for getting the coefficients
@variable(m, u[1:k+1])
@constraint(m, u[1]*0 + u[2]*0 + u[3]*0 + u[4] == 0)
@objective(m, Min, sum((raw[:,2] - A*u).^2))
```

```julia
status = solve(m)
uopt = getvalue(u)
println(status)
```

```julia
npts = 1000
xfine = linspace(0,10,npts)
ffine = ones(npts)
for j = 1:k
    ffine = [ffine.*xfine ones(npts)]
end
yfine = ffine * uopt
figure(figsize=(8,4))
plot(raw[:,1], raw[:,2],"r.", markersize = 5)
plot(xfine, yfine, "b-")
axis([0,10,0,1])
grid("on")
```

##### Part (a)

```julia
# Load the data file (ref: Boyd/263)
raw = readcsv("uy_data.csv");
u = raw[:,1];
y = raw[:,2];
T = length(u)
```

```julia

using PyPlot
figure(figsize=(12,4))
plot([u y], ".-")
legend(["input u","output y"], loc = "lower right")
```

```julia
# generate A matrix. Using more width creates better fit.  (MA model)
width = 5
A = zeros(T,width)
for i = 1:width
    A[i:end,i] = u[1:end-i+1]
end
wopt = A\y
yest_ma = A*wopt

A = zeros(T, width)
for i = 1:width
    A[i+1:end,i] = u[1:end-i] 
end
wopt = A\y
yest_ar = A*wopt

figure(figsize=(12,4))
plot(y,"g.-",yest_ma,"m.-",yest_ar,"c.-")
legend(["true output", "MA output","AR output"], loc="upper right");
title("Moving average model and Autoregressive model");
println()
println("Error for MA = ",norm(yest_ma-y))
println("Error for AR = ",norm(yest_ar-y))
```

##### Part (b)

```julia
# we use 2x the number of rows in A because 
# in ARMA, we have to fill the first half of
# rows with u values that work for AR
# and the second half for values related to
# MA
A = zeros(T, width*2)


# fill in values for AR
for i = 1:width
    A[i+1:end,i] = u[1:end-i] 
end

# fill in values for MA
for i = 1:width
    A[i:end,i+width] = u[1:end-i+1]
end

wopt = A\y
yest_arma = A*wopt

figure(figsize=(12,4))
plot(y,"g.-",yest_arma,"m.-")
legend(["true output", "ARMA output"], loc="upper right");
title("Autoregressive Moving Average Model");
println()
println("Error for ARMA = ",norm(yest_arma-y))
```
