```julia
using PyPlot
```

```julia

raw = readcsv("xy_data.csv")
figure(figsize=(8,4))
plot(raw[:,1], raw[:,2],"r.",markersize = 5)
axis([0,10,0,1])
grid("on")
```

```julia
# order of the polynomial to use
k = 3
```

```julia
# fit using a function of the form f(x) = u1 x^k + u2 x^(k-1) + ... + uk x + u{k+1}
n = length(raw[:,1])

# create the A matrix with coefficients
# the main objective here is to find the coefficients
# doing raw[i,1]^(k+1-j) is simply like doing x^i.
A = zeros(n,k+1)
for i = 1:n
    for j = 1:k+1
        A[i,j] = raw[i,1]^(k+1-j)
    end
end
```

```julia
# use Gurobi and Julia
using JuMP, Gurobi
```

```julia
m = Model(solver=GurobiSolver(OutputFlag=0))

# declare variable for getting the coefficients
@variable(m, u[1:k+1])
@constraint(m, u[1]*0 + u[2]*0 + u[3]*0 + u[4] == 0)
@objective(m, Min, sum((raw[:,2] - A*u).^2))
```

```julia
status = solve(m)
uopt = getvalue(u)
println(status)
```

```julia
npts = 1000
xfine = linspace(0,10,npts)
ffine = ones(npts)
for j = 1:k
    ffine = [ffine.*xfine ones(npts)]
end
yfine = ffine * uopt
figure(figsize=(8,4))
plot(raw[:,1], raw[:,2],"r.", markersize = 5)
plot(xfine, yfine, "b-")
axis([0,10,0,1])
grid("on")
```

##### Part (b)

```julia
# # obtain the position where we split the matrix
split_pos = 0
for i in 1:length(raw[:,1])
    if raw[i,1] < 4
        split_pos = i
    end
end


k = 3

# fit using a function of the form f(x) = u1 x^k + u2 x^(k-1) + ... + uk x + u{k+1}
n = length(raw[:,1])

# create the A matrix with coefficients
# the main objective here is to find the coefficients
# doing raw[i,1]^(k+1-j) is simply like doing x^i.
A = zeros(n,k+1)
for i = 1:n
    for j = 1:k+1
        A[i,j] = raw[i,1]^(k+1-j)
    end
end


##################### MODEL FOR 0 <= X < 4 #########################

m = Model(solver=GurobiSolver(OutputFlag=0))

# declare variable for getting the coefficients
@variable(m, u[1:k+1])
@constraint(m, u[1]*0 + u[2]*0 + u[3]*0 + u[4] == 0)

# create the piecewise function such that it computes two different
# plots two different functions. Our objective is to minimize
# the sum of squares of two functions.
@objective(m, Min, sum((raw[1:split_pos,2] - A[1:split_pos,:]*u).^2) +
                sum((raw[split_pos+1:200,2] - A[split_pos+1:200,:]*u).^2))

status = solve(m)
uopt = getvalue(u)
println(status)

npts = 1000
xfine = linspace(0,10,npts)
ffine = ones(npts)
for j = 1:k
    ffine = [ffine.*xfine ones(npts)]
end
yfine = ffine * uopt
figure(figsize=(8,4))
plot(raw[:,1], raw[:,2],"r.", markersize = 5)
plot(xfine[1:400], yfine[1:400], "b-")
plot(xfine[400+1:1000], yfine[400+1:1000], "g-")

axis([0,10,0,1])
grid("on")


```
